```python
Train_datadict={}
train_datadict={}

for i in range(26):
    x_train=datasets[i].iloc[:,1:]
    x_train=x_train.T
    y_train=datasets[i].iloc[:,0]
    y_train=np.array([y_train])
    x_test=Datasets.iloc[:,1:]
    x_test=x_test.T
    y_test=Datasets[i].iloc[:,0]
    y_test=np.array([y_test])
    Train_datadict['D'+str(i)]=[x_train,y_train]
    train_datadict['D'+str(i)]=[x_test,y_test]
```

```python
itterations = 10000000
alpha = 0.160 
m = 88800
cost_values = []
trained_parameters = []
for trainsets in range(26):
    X=Train_datadict['D'+str(trainsets)][0]
    Y=Train_datadict['D'+str(trainsets)][1]
    weights=np.random.randn(1,784)
    bias=0
    costfunc=[]
    f=0
    print('training for dataset'+str(trainsets))
    for i in range(1,itterations+1):
        #logistic function
        z = np.dot(weights,X) + bias 
        hypothesis = 1/(1 + np.exp(-z)) 
        #cost function
        j = 1/m*(-1*(np.sum(Y*np.log(hypothesis+10**(-7)) + (1-Y)*np.log(1-hypothesis+10**(-7)))))
        costfunc.append(j)
        f+=1
        #gradient decent
        dw =  1/m * np.dot(hypothesis-Y,X.T)
        db =  1/m * np.sum(hypothesis-Y)
        weights = weights - alpha*dw
        bias = bias - alpha*db
        #stop training
        if i%2000 == 0:
            print('running @ ',j)
        if i%2 == 0:
            if abs(j-costfunc[-2])<0.000001:
                if abs(j-costfunc[-3])<0.000001:
                    break     
    cost_values.append(costfunc)  
    trained_parameters.append([weights,bias])      
    print('itteration number:',f)
    
    
    
```

```python
#accuracy(train)
for datasetnum in range(26):
    X=Train_datadict['D'+str(datasetnum)][0]
    Y=Train_datadict['D'+str(datasetnum)][1]
    weights=trained_parameters[datasetnum][0]
    bias=trained_parameters[datasetnum][1]
    correct=0
    for i in range(1000):
        z=np.dot(weights,X.T.iloc[i,:])+bias
        hypothesis = 1/(1 + np.exp(-z))
        if np.logical_and(hypothesis >= 0.5,Y.T[i,0]==1):
            correct+=1
        if np.logical_and(hypothesis < 0.5,Y.T[i,0]==0):
            correct+=1
    acc=(correct/1000)*100
    print('accuracy for dataset'+str(datasetnum),"=",acc)
    
```
